# IMPORTING MODULES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv("PROJECT3_CSV.csv")
```

```python
df.head()      #active -- total,discharged, deaths
```

```python
df.info()
```

# GRAPHS


```python
df.plot.area();
```

# RELATION BETWEEN TOTAL CASES  AND ACTIVE 

```python
df.groupby('Total Cases')['Active'].value_counts().unstack('Active').plot(kind='barh', stacked=True,
                                                                    figsize=(15, 15),colormap = "Accent");
```

# RELATION BETWEEN  ACTIVE CASES  AND DEATHS

```python
df.groupby('Active')['Deaths'].value_counts().unstack('Deaths').plot(kind='barh', stacked=True,
                                                                    figsize=(15, 15),colormap = "coolwarm");
```

# RELATION BETWEEN STATE/UT'S  AND ACTIVE  CASES 

```python
df.groupby('State/UTs')['Active'].value_counts().unstack('Active').plot(kind='barh', stacked=True,
                                                                    figsize=(15, 15),colormap = "plasma");
```
